# Task 4: Model Building & Comparison


```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import RandomForestClassifier

from sklearn.ensemble import GradientBoostingClassifier

from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Samples :", X_train.shape)

print("Testing Samples :", X_test.shape)
```

```python
print(y_train.value_counts())

print()

print(y_test.value_counts())
```

```python
log_model = LogisticRegression(
    class_weight="balanced",
    random_state=42,
    max_iter=1000
)

log_model.fit(X_train,y_train)
```

```python
log_pred = log_model.predict(X_test)
```

```python
log_accuracy = accuracy_score(
    y_test,
    log_pred
)

print("Logistic Regression Accuracy :",log_accuracy)
```

```python
rf_model = RandomForestClassifier(

    n_estimators=300,

    random_state=42,

    class_weight="balanced"

)

rf_model.fit(X_train,y_train)
```

```python
rf_pred = rf_model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(
    y_test,
    rf_pred
)

print("Random Forest Accuracy :",rf_accuracy)
```

```python
gb_model = GradientBoostingClassifier(
    random_state=42
)

gb_model.fit(X_train,y_train)
```

```python
gb_pred = gb_model.predict(X_test)
```

```python
gb_accuracy = accuracy_score(
    y_test,
    gb_pred
)

print("Gradient Boosting Accuracy :",gb_accuracy)
```

```python
log_prob = log_model.predict_proba(X_test)[:,1]

rf_prob = rf_model.predict_proba(X_test)[:,1]

gb_prob = gb_model.predict_proba(X_test)[:,1]
```

```python
results = pd.DataFrame({

    "Model":[

        "Logistic Regression",

        "Random Forest",

        "Gradient Boosting"

    ],

    "Accuracy":[

        log_accuracy,

        rf_accuracy,

        gb_accuracy

    ]

})

results
```

```python
results.sort_values(
    by="Accuracy",
    ascending=False
)
```

```python
best_model = results.sort_values(
    by="Accuracy",
    ascending=False
)

best_model
```

```python
print(best_model.iloc[0])
```

```python
print("Logistic :",log_model.score(X_train,y_train))

print("Random Forest :",rf_model.score(X_train,y_train))

print("Gradient Boost :",gb_model.score(X_train,y_train))
```

```python
print("Logistic :",log_model.score(X_test,y_test))

print("Random Forest :",rf_model.score(X_test,y_test))

print("Gradient Boost :",gb_model.score(X_test,y_test))
```

```python
comparison = pd.DataFrame({

    "Model":[

        "Logistic",

        "Random Forest",

        "Gradient Boost"

    ],

    "Training Accuracy":[

        log_model.score(X_train,y_train),

        rf_model.score(X_train,y_train),

        gb_model.score(X_train,y_train)

    ],

    "Testing Accuracy":[

        log_model.score(X_test,y_test),

        rf_model.score(X_test,y_test),

        gb_model.score(X_test,y_test)

    ]

})

comparison
```

# Task 5: Model Evaluation


```python
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_auc_score,
    roc_curve,
    precision_score,
    recall_score,
    f1_score
)
```

```python
log_precision = precision_score(y_test, log_pred)

log_recall = recall_score(y_test, log_pred)

log_f1 = f1_score(y_test, log_pred)

log_auc = roc_auc_score(y_test, log_prob)

print("Precision :",log_precision)

print("Recall :",log_recall)

print("F1 Score :",log_f1)

print("ROC-AUC :",log_auc)
```

```python
print(classification_report(y_test,log_pred))
```

```python
cm_log = confusion_matrix(y_test,log_pred)

cm_log
```

```python
disp = ConfusionMatrixDisplay(
    confusion_matrix=cm_log,
    display_labels=["Stayed","Left"]
)

disp.plot(cmap="Blues")

plt.title("Logistic Regression Confusion Matrix")

plt.show()
```

```python
rf_precision = precision_score(y_test,rf_pred)

rf_recall = recall_score(y_test,rf_pred)

rf_f1 = f1_score(y_test,rf_pred)

rf_auc = roc_auc_score(y_test,rf_prob)

print("Precision :",rf_precision)

print("Recall :",rf_recall)

print("F1 :",rf_f1)

print("ROC-AUC :",rf_auc)
```

```python
gb_precision = precision_score(y_test,gb_pred)

gb_recall = recall_score(y_test,gb_pred)

gb_f1 = f1_score(y_test,gb_pred)

gb_auc = roc_auc_score(y_test,gb_prob)

print("Precision :",gb_precision)

print("Recall :",gb_recall)

print("F1 :",gb_f1)

print("ROC-AUC :",gb_auc)
```

# Task 7: HR Insights & Business Recommendations

# Question 1
Which three factors most strongly predict employee attrition?

### Answer

Based on the feature importance analysis of the Gradient Boosting model, the three strongest predictors of employee attrition are:

1. **OverTime**
   Employees who regularly work overtime are significantly more likely to leave the company.

2. **Monthly Income**
   Employees with comparatively lower salaries show a higher likelihood of resignation.

3. **Years at Company**
   Employees are more likely to leave during their early years in the organization before becoming long-term employees.

Other important factors include Job Satisfaction, Environment Satisfaction, Distance From Home, Age, and Stock Option Level.

# Question 2
Which department or job role should HR prioritize?

### Answer

The Sales department experiences the highest employee attrition rate.

Among job roles, Sales Representatives and Laboratory Technicians have the highest turnover.

HR should prioritize these departments by conducting employee engagement programs, career development initiatives, and regular retention discussions.

# Question 3
Does salary alone explain attrition?

### Answer

No.

Although employees with lower salaries tend to leave more frequently, salary alone does not explain employee attrition.

The analysis shows that overtime, work-life balance, job satisfaction, years at the company, and career progression also contribute significantly to employee turnover.

Therefore, employee attrition is influenced by multiple organizational and personal factors rather than salary alone.

# Question 4
Two HR Recommendations

### Recommendation 1

Develop a proactive employee retention program for newly hired employees during their first three years.

This program should include mentorship, career planning, regular feedback sessions, and employee engagement activities.

---

### Recommendation 2

Reduce excessive overtime by improving workload distribution and encouraging better work-life balance.

Employees with consistently high overtime should receive regular well-being check-ins and workload assessments.

# Question 5
Model Limitation

### Limitation

Although the machine learning model predicts employee attrition effectively, it has several limitations.

The model is trained on historical employee records from a single company and may not generalize perfectly to other organizations.

Additionally, the dataset does not include factors such as employee mental health, organizational culture, leadership quality, or external job market conditions, which can also influence attrition decisions.

Therefore, HR should use the model as a decision-support tool rather than replacing human judgment.
